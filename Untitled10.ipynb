```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/youtube.csv')

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Display dataset
print(df.head())

# Dataset information
print(df.info())

# Statistical summary
print(df.describe())

# Check missing values
print(df.isnull().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing values
df = df.fillna(0)

# Check available columns
print(df.columns)

# Top 10 most viewed videos
top_views = df.sort_values(by='views', ascending=False).head(10)

print(top_views[['title', 'views']])

# Count categories
category_count = df['category_id'].value_counts()

print(category_count)


plt.figure(figsize=(10,5))

sns.countplot(
    y='category_id',
    data=df,
    order=df['category_id'].value_counts().index
)

plt.title("Top Trending Categories")

plt.show()
plt.figure(figsize=(8,5))

sns.scatterplot(x='views', y='likes', data=df)

plt.title("Views vs Likes")
plt.show()
top_channels = df['channel_title'].value_counts().head(10)

top_channels.plot(kind='bar')

plt.title("Top Trending Channels")
plt.xlabel("Channel Name")
plt.ylabel("Count")

plt.show()
top_likes = df.sort_values(by='likes', ascending=False).head(10)

print(top_likes[['title', 'likes']])
top_comments = df.sort_values(by='comment_count', ascending=False).head(10)

print(top_comments[['title', 'comment_count']])
plt.figure(figsize=(8,5))

plt.hist(df['likes'], bins=20)

plt.title("Likes Distribution")
plt.xlabel("Likes")
plt.ylabel("Frequency")

plt.show()
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
